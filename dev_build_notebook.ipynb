```python
df_loc_codes = pd.read_csv(location_list_path)
df_loc_codes

# country_codes,hdx_endpoints,metadata_endpoints
```

```python
# start a basic logging set up
logging.basicConfig(level=logging.INFO, format='%(message)s')

# initialise a list to hold the response
hdx_3Ws_data = []

# request data for each country and push to a list
for location_code in df_loc_codes["country_codes"]:
    params = {
        "location_code": location_code,
        "limit": "1000"
    }
    headers = {
        "X-HDX-HAPI-APP-IDENTIFIER": "aGR4ZHc6bHBhZGdldHQyM0BnbWFpbC5jb20="
    }
    
    response = requests.get(f"https://hapi.humdata.org/api/v1/coordination-context/operational-presence", params=params, headers=headers)

    if response.status_code == 200: 
        data = response.json()
    
        # check if 'data' key exists and is not empty
        if 'data' in data and data['data']:  
            hdx_3Ws_data.append(data)
            logging.info(f"The call to location code {location_code} has returned data with {len(data['data'])} rows.")
        else:
            logging.info(f"The call to location code {location_code} did not return any data.")
    else:
        raise Exception("Extracting 3Ws data failed for location code {location_code}. Please check if limits have been reached")
```

```python
parsed_json = json.loads(hdx_3Ws_data[0])
print(parsed_json)
```

```python
df = pd.json_normalize(hdx_3Ws_data[6]['data'])
df
```

```python
dataframes = []

# loop through each json response in the list
for i in range(len(hdx_3Ws_data)):
    df = pd.json_normalize(hdx_3Ws_data[i]['data'])
    dataframes.append(df)

# concatenate all the dataframes into one
combined_df = pd.concat(dataframes, ignore_index=True)

print(combined_df)
```

```python
# above is data loading with pandas, you can either replace the data or append to it, or throw an error

# if you need upsert loads, --> it inserts or updates
# insert data if the record is new
# update data if teh record alreaady exists in the db

from sqlalchemy import Table, Column, Integer, String, MetaData, Float
from sqlalchemy.dialects import postgresql
```

```python
# create schema of table to create in the database

metadata = MetaData() # requirement of the lib

orgs3ws_table = Table(
    "coord3ws", # name for the table
    metadata,
    # now have to tell it the columns that you will need
    Column("location_ref", Integer, primary_key=False),
    Column("location_code", String, primary_key=False),
    Column("location_name", String, primary_key=False),
    Column("admin1_ref", Integer, primary_key=False),
    Column("admin1_code", String, primary_key=False),
    Column("admin1_name", String, primary_key=False),
    Column("admin2_ref", Integer, primary_key=False),
    Column("admin2_code", String, primary_key=False),
    Column("admin2_name", String, primary_key=False),
    Column("resource_hdx_id", String, primary_key=False),
    Column("org_acronym", String, primary_key=False),
    Column("org_name", String, primary_key=False)
    Column("sector_code", String, primary_key=False),
    Column("sector_name", String, primary_key=False),
    Column("reference_period_start", String, primary_key=False),
    Column("reference_period_end", String, primary_key=False),
    Column("org_type_code", String, primary_key=False),
    Column("org_type_description", String, primary_key=False)
)
```
